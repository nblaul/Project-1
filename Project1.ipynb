```python
import requests
import pandas as pd
from pprint import pprint

url = 'http://spreadtrends.com/api/nfl/games/2020?format=json'

game_data = requests.get(url).json()

#create empty lists
temperature = []
conditions = []
windspeed = []
game_id = []
season = []
surface = []
precipitation = []
game_type = []

for x in range(0, (len(game_data['results']))):
    try:
        game_id.append(game_data['results'][x]['gameId'])
        season.append(game_data['results'][x]['season'])
        temperature.append(game_data['results'][x]['temperature'])
        conditions.append(game_data['results'][x]['conditions'])
        windspeed.append(game_data['results'][x]['windSpeed'])
        surface.append(game_data['results'][x]['surface'])
        game_type.append(game_data['results'][x]['type'])
        precipitation.append(game_data['results'][x]['precipitation'])
    except KeyError:
        #print(game_data['results'][x]['gameId'])
        temperature.append(None)
        season.append(None)
        conditions.append(None)
        windspeed.append(None)
        surface.append(None)
        precipitation.append(None)
    
#weather_df = pd.DataFrame({'GameID': game_id,
                    #'Temperature': temperature,
                    #'Conditions': conditions,
                    #'Windspeed': windspeed,
                    #'Season': season,
                    #'Surface': surface,
                    #'Game_Type': game_type})
                           
#weather_df = weather_df.loc[weather_df['Game_Type'] == 'Regular']
#weather_df.reset_index()

#weather_df
print(len(game_id))
print(len(temperature))
print(len(season))
print(len(conditions))
print(len(windspeed))
print(len(surface))
print(len(precipitation))
```
